```python
import nltk
nltk.download ('punkt')
sentence_data = "The first sentence is about python. the second: about django. You can learn python, django adn data analysis here."
nltk_tokens = nltk.sent_tokenize(sentence_data)
print(nltk_tokens)
```

```python
german_tokenizer = nltk.data.load('tokenizers/punkt/german.pickle')
german_tokens = german_tokenizer.tokenize('suck a nigga dick? suck a nigga dick??')
print(german_tokens)

```

```python
word_data = 'I love hotdog because it is not a dog, we love hotdog because it is hotdog and not a dog. we all love hotdog because it is yummy'
nltk_tokens = nltk.word_tokenize(word_data)
print(nltk_tokens)

```
